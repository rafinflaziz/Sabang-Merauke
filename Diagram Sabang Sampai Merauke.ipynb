```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
with plt.xkcd():
    plt.rcParams.update({'font.size': 11})
    fig = plt.figure(figsize=(13, 6))
    ax = fig.add_axes((0.1, 0.1, 0.8, 0.8))

    # ============ LYRICS BOX AND TEXT ============ #

    # both 1
    both_1 = patches.Rectangle(
        (0.04, 0.45), 0.23, 0.1, alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.05, 0.49, 'DARI SABANG SAMPAI MERAUKE')

    # ori 1
    ori_1 = patches.Rectangle((0.29, 0.85), 0.19, 0.1,
                              alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.3, 0.89, 'BERJAJAR PULAU-PULAU')

    # ori 2
    ori_2 = patches.Rectangle((0.57, 0.85), 0.28, 0.1,
                              alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.58, 0.89, 'SAMBUNG MENYAMBUNG MENJADI SATU')

    # ori 3
    ori_3 = patches.Rectangle((0.46, 0.65), 0.14, 0.1,
                              alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.47, 0.69, 'ITULAH INDONESIA')

    # both 2
    both_2 = patches.Rectangle(
        (0.44, 0.45), 0.18, 0.1, alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.45, 0.49, 'INDONESIA TANAH AIRKU')

    # ori 4
    ori_4 = patches.Rectangle((0.69, 0.65), 0.17, 0.1,
                              alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.70, 0.69, 'AKU BERJANJI PADAMU')

    # ori 5
    ori_5 = patches.Rectangle((0.93, 0.65), 0.04, 0.1,
                              alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.94, 0.69, '. . .')

    # indomie 1
    indomie_1 = patches.Rectangle(
        (0.37, 0.23), 0.25, 0.1, alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.38, 0.27, 'DARI MIANGAS SAMPAI PULAU ROTE')

    # indomie 2
    indomie_2 = patches.Rectangle(
        (0.69, 0.23), 0.152, 0.1, alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.70, 0.27, 'INDOMIE SELERAKU')

    # alternatif
    alternatif = patches.Rectangle(
        (0.04, 0.06), 0.23, 0.1, alpha=1, edgecolor='black', facecolor='none')
    ax.text(0.05, 0.1, 'DARI TIMOR SAMPAI KE TALAUD')

    ax.text(0.01, 0.28, 'PRE-REVISED', color='tab:blue')

    ax.add_patch(ori_1)
    ax.add_patch(ori_2)
    ax.add_patch(ori_3)
    ax.add_patch(ori_4)
    ax.add_patch(ori_5)
    ax.add_patch(indomie_1)
    ax.add_patch(indomie_2)
    ax.add_patch(both_1)
    ax.add_patch(both_2)
    ax.add_patch(alternatif)
    
    # ============ ARROWS ============ #

    # Define the coordinates for the starting and ending points of the arrows
    arrow_start = [(0.479, 0.9), (0.568, 0.88),
                   (0.53, 0.645), (0.862, 0.7),
                   (0.49, 0.337), (0.18, 0.165)]
    arrow_end = [(0.57, 0.9), (0.53, 0.75),
                 (0.53, 0.55), (0.929, 0.7),
                 (0.531, 0.44), (0.44, 0.48)]

    dotted_arrow_start = [(0.17, 0.554), (0.17, 0.441),
                          (0.17, 0.441), (0.622, 0.51), (0.622, 0.50)]
    dotted_arrow_end = [(0.29, 0.9), (0.37, 0.264),
                        (0.17, 0.165), (0.689, 0.7), (0.69, 0.27)]

    # Create and add arrows to the figure
    for start, end in zip(arrow_start, arrow_end):
        arrow = patches.FancyArrowPatch(
            start, end,
            arrowstyle='-|>',
            mutation_scale=15,
            color='black',
        )
        ax.add_patch(arrow)

    for start, end in zip(dotted_arrow_start, dotted_arrow_end):
        arrow = patches.FancyArrowPatch(
            start, end,
            arrowstyle='-|>',
            mutation_scale=15,
            color='black',
            linestyle='--'
        )
        ax.add_patch(arrow)
        
    # ============ DOTTED LINES ============ #

    plt.plot([0, 0.035], [0.5, 0.5], linestyle='--', color='tab:red')
    plt.plot([0.275, 0.436], [0.5, 0.5], linestyle='--', color='tab:red')
    plt.plot([0.625, 1], [0.5, 0.5], linestyle='--', color='tab:red')
    plt.plot([0, 0.3], [0.25, 0.25], linestyle='--', color='tab:blue')
    plt.vlines(x=0.3, linestyle='--', ymin=0, ymax=0.25)
    
    
    # ============ FINAL MARKUP ============ #
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.title('Diagram lirik "Dari Sabang Sampai Merauke" vs "Jingle Indomie"')

    # Save and show the figure
    plt.savefig('indomie.png', dpi=150)
    plt.show()
```
